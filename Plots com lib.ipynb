```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Definindo variaveis e contando valores e suas ocorrencias
df = pd.read_csv(r'/home/braulio/Documents/python-master/2019-05-022.csv', delimiter=',', low_memory=False)
avg = df['Average'].sort_values(ascending=True)
counter = dict()
totNumbers = len(avg)

for n in avg:
    try:
        counter[n] +=1
    except KeyError:
        counter[n] = 1

cdfK = []
cdfV = []
pdfV = []
rawValue = []
lastVal = 0
for key,value in counter.items():
    lastVal += (value/totNumbers)
    rawValue.append(key)
    cdfK.append(key)
    cdfV.append(lastVal)
```

```python
#CDF utilizando Biblioteca Seaborn
sns.kdeplot(cdfK,cumulative=True)
plt.show()
```

```python
#PDF utilizando biblioteca seaborn
sns.distplot(cdfK, hist=False)
plt.show()
```
